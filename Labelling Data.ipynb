```python
def label_text(text):
    if not isinstance(text, str):
        return "Non_UK"
    text = text.lower()
    labels = set()
    for label, words in keywords.items():
        if any(word in text for word in words):
            labels.add(label)
    return ", ".join(sorted(labels)) if labels else "Non_UK"
```

```python
df['label_UK'] = df['full_text'].fillna("").apply(label_text)
```

```python
output_file = "final wanita dataset.csv"
df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"File disimpan sebagai: {output_file}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words_ind = stopwords.words('indonesian')

df_tfidf = df[['full_text', 'label_UK']].dropna()

labels = ['UK_victim', 'UK_mockery', 'UK_misogyny', 'UK_patriarki', 'Non_UK']
docs_by_label = {
    label: df_tfidf[df_tfidf['label_UK'].str.contains(label, na=False)]['full_text'].tolist()
    for label in labels
}

top_keywords_per_label = {}
for label, docs in docs_by_label.items():
    if len(docs) < 2:
        continue
    vectorizer = TfidfVectorizer(stop_words=stop_words_ind, max_features=1000)
    tfidf_matrix = vectorizer.fit_transform(docs)
    mean_tfidf = tfidf_matrix.mean(axis=0).A1
    terms = vectorizer.get_feature_names_out()
    tfidf_scores = dict(zip(terms, mean_tfidf))
    top_keywords = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:30]
    top_keywords_per_label[label] = top_keywords
```

```python
for label, keywords in top_keywords_per_label.items():
    print(f"\nTop keywords untuk label: {label}")
    for word, score in keywords:
        print(f"{word}: {score:.4f}")
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(keywords))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f'WordCloud untuk {label}', fontsize=16)
plt.show()
```
