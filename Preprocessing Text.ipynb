```python
import re

def clean_twitter_text(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'[^A-Za-z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_text'] = df['full_text'].astype(str).apply(clean_twitter_text)
```

Preprocessing Text
1. Case Folding
2. Tokenization
3. Stopword Removal
4. Stemming

```python
# Case folding
df['case_folding'] = df['full_text'].str.lower()
df
```

```python
import re

def clean_twitter_text(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'[^A-Za-z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_text'] = df['case_folding'].astype(str).apply(clean_twitter_text)
df
```

```python
# Tokenisasi

df['tokenize'] = df['clean_text'].apply(lambda x: x.split())
df
```

```python
stop_words = {
    'yang', 'dan', 'di', 'ke', 'dari', 'ini', 'itu', 'untuk', 'dengan', 'pada',
    'sebagai', 'adalah', 'tidak', 'atau', 'karena', 'juga', 'agar', 'oleh',
    'sudah', 'saja', 'sangat', 'masih', 'lebih', 'kurang', 'bisa', 'dalam',
    'apa', 'iya', 'jadi', 'kalau', 'kalo', 'nanti', 'lagi', 'semua',
      'udah', 'aja', 'kayak', 'kayaknya', 'gitu', 'ga', 'nggak', 'ngga', 'kok',
    'loh', 'lah', 'dong', 'deh', 'si', 'eh', 'nih', 'ya', 'ye', 'banget', 'amat',
    'emang', 'doang', 'cuma', 'gue', 'gua', 'lu', 'loe', 'elo', 'kita', 'kami',
    'kenapa', 'mengapa', 'siapa', 'dimana', 'kapan', 'berapa'
}

df['stopword_removal'] = df['tokenize'].apply(
    lambda tokens: [word for word in tokens if word not in stop_words]
)

df
```
